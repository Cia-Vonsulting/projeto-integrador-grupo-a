Bibliotecas necessárias

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

Url do site que coletaremos os dados

```python
url = "https://www.nyc.gov/site/tlc/about/tlc-trip-record-data.page"
```

Configurando agente para realizar a requisição ao site

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}
```

Acessando página com links

```python
response = requests.get(url, headers=headers)
response.raise_for_status()
```

```python
soup = BeautifulSoup(response.text, features='html.parser')
```

```python
base_url = "https://www.nyc.gov"
parquet_links = []
```

Pegando todos os links que terminando com parquet

```python
for link in soup.find_all('a', href=True):
    href = link['href']
    if href.endswith('.parquet'):
        parquet_links.append(href)
```

```python
yellow_tripdata_links = list(filter(lambda href: 'yellow_tripdata' in href, parquet_links))
```

```python
sample_url = yellow_tripdata_links[0]
```

```python
sample_url
```

```python
response_download = requests.get(sample_url)
```

```python
file_Path = '../dataset/nyc-tlc/2024/test.parquet'
```

```python
if response_download.status_code == 200:
    with open(file_Path, 'wb') as file:
        file.write(response_download.content)
    print('File downloaded successfully')
else:
    print('Failed to download file')
```

```python
test_parquet = pd.read_parquet(file_Path)
```
